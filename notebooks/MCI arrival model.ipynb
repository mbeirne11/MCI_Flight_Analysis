```python
MCI_origin_df=MCI_flights_df.loc[MCI_flights_df['ORIGIN'] == 'MCI']
MCI_dest_df=MCI_flights_df.loc[MCI_flights_df['DEST'] == 'MCI']
```

```python
X_dest = MCI_dest_df[['MONTH','DAY_OF_MONTH', 'DAY_OF_WEEK', 'MKT_UNIQUE_CARRIER', 'ORIGIN_CITY_NAME', 'ORIGIN_STATE_NM','ORIGIN', 'CRS_DEP_TIME']]
```

```python
y_dest = MCI_dest_df[['Status']]
```

```python
months = []
days = []
month_dict = {1:'January', 2: 'February',3:'March', 4: 'April',5:'May', 6: 'Jun',7:'July', 8: 'August',9:'September', 10: 'October',11:'November', 12: 'December'}
for row in X_dest.iterrows():
    month = month_dict[row[1]['MONTH']]
    day = f"{row[1]['DAY_OF_MONTH']}"
    days.append(day)
    months.append(month)
```

```python
X_dest['MONTH'] = months
X_dest['DAY_OF_MONTH'] = days
```

```python
dows = []
for row in X_dest.iterrows():
    if row[1]['DAY_OF_WEEK'] == 1:
        dow = 'Monday'
    if row[1]['DAY_OF_WEEK'] == 2:
        dow = 'Tuesday'
    if row[1]['DAY_OF_WEEK'] == 3:
        dow = 'Wednesday'
    if row[1]['DAY_OF_WEEK'] == 4:
        dow = 'Thursday'
    if row[1]['DAY_OF_WEEK'] == 5:
        dow = 'Friday'
    if row[1]['DAY_OF_WEEK'] == 6:
        dow = 'Saturday'
    if row[1]['DAY_OF_WEEK'] == 7:
        dow = 'Sunday'
    dows.append(dow)
```

```python
X_dest['DAY_OF_WEEK'] = dows
X_dest
```

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from keras.layers import Dense, Dropout
```

```python
X_dest_dummies = pd.get_dummies(X_dest,drop_first=True)
X_dest_dummies
```

```python
scaler = StandardScaler()
scaler.fit(X_dest_dummies)
transformed_data = scaler.transform(X_dest_dummies)
# Creating a DataFrame with with the scaled data
transformed_data_df = pd.DataFrame(transformed_data)

# Display sample data
transformed_data_df
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
pca = PCA().fit(transformed_data_df)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```

```python
from sklearn.decomposition import PCA
# Create a PCA model
nc = 80
pca = PCA(n_components=nc)
pca_data = pca.fit_transform(transformed_data_df)
pca_dict = {}
for i in range(0,nc):
    pca_dict[f'pca{i+1}']=pca_data[:,i]
# Display sample data
pca_df = pd.DataFrame(pca_dict)
pca_df
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X_dest_dummies, y_dest['Status_On Time'], random_state=78)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn_model = tf.keras.models.Sequential()

# First hidden layer
nn_model.add(tf.keras.layers.Dense(units=64, activation="relu", input_dim=len(pca_df.columns)))
# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=128, activation="relu"))
nn_model.add(Dropout(0.1))
# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=512, activation="relu"))
nn_model.add(Dropout(0.3))
# Third hidden layer
nn_model.add(tf.keras.layers.Dense(units=1024, activation="relu"))
nn_model.add(Dropout(0.5))
# Third hidden layer
nn_model.add(tf.keras.layers.Dense(units=512, activation="sigmoid"))
nn_model.add(Dropout(0.3))
nn_model.add(tf.keras.layers.Dense(units=64, activation="sigmoid"))
nn_model.add(Dropout(0.1))
# Output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn_model.fit(X_train, y_train, epochs=10)
```

```python
model_loss, model_accuracy = nn_model.evaluate(X_test,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
