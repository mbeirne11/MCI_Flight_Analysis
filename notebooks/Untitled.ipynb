```python
import pandas as pd
```

```python
df = pd.read_csv('../Resources/flight_data.csv').drop(['Unnamed: 0'],axis=1)
df
```

```python
df.columns
```

```python
months_df = df.groupby(['MONTH'])['Status'].describe()
years_df = df.groupby(['YEAR'])['Status'].describe()
dows_df = df.groupby(['DAY_OF_WEEK'])['Status'].describe()
origins_df = df.groupby(['ORIGIN'])['Status'].describe()
dests_df = df.groupby(['DEST'])['Status'].describe()
carriers_df = df.groupby(['MKT_UNIQUE_CARRIER'])['Status'].describe()
```

```python
years_df.to_csv('../Resources/stats_by_year.csv')
months_df.to_csv('../Resources/stats_by_month.csv')
dows_df.to_csv('../Resources/stats_by_dow.csv')
origins_df.to_csv('../Resources/stats_by_origin.csv')
dests_df.to_csv('../Resources/stats_by_dest.csv')
carriers_df.to_csv('../Resources/stats_by_carrier.csv')
```
