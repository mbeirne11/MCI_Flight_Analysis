```python
cols = ['YEAR', 'MONTH', 'DAY_OF_MONTH', 'TAIL_NUM','MKT_CARRIER_FL_NUM','DEP_TIME_BLK']
more_cols = ['YEAR', 'MONTH', 'DAY_OF_MONTH',  'MKT_UNIQUE_CARRIER','TAIL_NUM','DEP_TIME_BLK','ORIGIN','DEST','DEP_DELAY_NEW','ARR_DELAY_NEW']
MCI_flights_df.loc[(MCI_flights_df['MKT_UNIQUE_CARRIER'] == 'WN')].sort_values(by = cols)[more_cols].head(50)
```

```python
MCI_origin_df=MCI_flights_df.loc[MCI_flights_df['ORIGIN'] == 'MCI']
MCI_dest_df=MCI_flights_df.loc[MCI_flights_df['DEST'] == 'MCI']
```

```python
tails = []
for row in MCI_origin_df.iterrows():
    tail_group = (row[1]['TAIL_NUM'][-1])
    tails.append(tail_group)
```

```python
MCI_origin_df['TAIL_NUM'] = tails
```

```python
MCI_origin_df['TAIL_NUM'].unique()
```

```python
avg_arr_delay_dict = MCI_dest_df.groupby(['MONTH', 'DAY_OF_WEEK', 'MKT_UNIQUE_CARRIER', 'DEP_TIME_BLK'])['ARR_DELAY_NEW'].mean()
avg_arr_delay_dict
```

```python
avg_arr_delays = []
for row in MCI_origin_df.iterrows():
    month = row[1]['MONTH']
    dow = row[1]['DAY_OF_WEEK']
    carrier = row[1]['MKT_UNIQUE_CARRIER']
    time = row[1]['DEP_TIME_BLK']
    try:
        avg_arr_delay = avg_arr_delay_dict[month][dow][carrier][time]
        avg_arr_delays.append(avg_arr_delay)
    except:
        avg_arr_delays.append(12)
            
```

```python
MCI_origin_df['AVG_ARR_DELAY']=avg_arr_delays
```

```python
avg_arr_delay_blks = []
for row in X_origin.iterrows():
    avg_arr_delay = row[1]['AVG_ARR_DELAY'] 
    if avg_arr_delay <= 5.300000:
        avg_arr_delay_blk = '0-25%'
    elif avg_arr_delay <= 10.142857:
        avg_arr_delay_blk = '25-50%'
    elif avg_arr_delay <= 18.612245:
        avg_arr_delay_blk = '50-75%'
    else:
        avg_arr_delay_blk = '75-100%'
    avg_arr_delay_blks.append(avg_arr_delay_blk)
```

```python
months = []
days = []
month_dict = {1:'January', 2: 'February',3:'March', 4: 'April',5:'May', 6: 'Jun',7:'July', 8: 'August',9:'September', 10: 'October',11:'November', 12: 'December'}
for row in X_origin.iterrows():
    month = month_dict[row[1]['MONTH']]
    day = f"{row[1]['DAY_OF_MONTH']}"
    days.append(day)
    months.append(month)
```

```python
X_origin['MONTH'] = months
X_origin['DAY_OF_MONTH'] = days
```

```python
dows = []
for row in X_origin.iterrows():
    if row[1]['DAY_OF_WEEK'] == 1:
        dow = 'Monday'
    if row[1]['DAY_OF_WEEK'] == 2:
        dow = 'Tuesday'
    if row[1]['DAY_OF_WEEK'] == 3:
        dow = 'Wednesday'
    if row[1]['DAY_OF_WEEK'] == 4:
        dow = 'Thursday'
    if row[1]['DAY_OF_WEEK'] == 5:
        dow = 'Friday'
    if row[1]['DAY_OF_WEEK'] == 6:
        dow = 'Saturday'
    if row[1]['DAY_OF_WEEK'] == 7:
        dow = 'Sunday'
    dows.append(dow)
```

```python
X_origin['DAY_OF_WEEK'] = dows
X_origin
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X_origin_dummies, y_origin_dummies['Status_On Time'], random_state=78)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn_model = tf.keras.models.Sequential()

# First hidden layer
nn_model.add(tf.keras.layers.Dense(units=64, activation="relu", input_dim=len(X_origin_dummies.columns)))
# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=128, activation="relu"))
nn_model.add(Dropout(0.2))
# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=512, activation="relu"))
nn_model.add(Dropout(0.5))
# Third hidden layer
nn_model.add(tf.keras.layers.Dense(units=1024, activation="relu"))
nn_model.add(Dropout(0.7))
# Third hidden layer
nn_model.add(tf.keras.layers.Dense(units=512, activation="sigmoid"))
nn_model.add(Dropout(0.5))
nn_model.add(tf.keras.layers.Dense(units=64, activation="sigmoid"))
nn_model.add(Dropout(0.2))
# Output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn_model.fit(X_train, y_train, epochs=10)
```

```python
model_loss, model_accuracy = nn_model.evaluate(X_test,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
