```python
# Dependencies
import json
import requests
import pandas as pd

from config import api_key
```

```python
df = pd.read_csv('../Resources/MCI_21-23.csv').drop(['Unnamed: 0'],axis=1)
df
```

```python
cities = list(df['ORIGIN_CITY_NAME'].unique())
cities
```

```python
responses = []
for city in cities:
    if '/' in city:
        city = city.split('/')[1]
    url = f'http://api.openweathermap.org/geo/1.0/direct?q={city}&limit={1}&appid={api_key}'
    response = requests.get(url).json()
    responses.append(response)
```

```python
for r in responses:
    print(f"{r[0]['name']},{r[0]['lat']},{r[0]['lon']}")
```

```python
city_coords = []
for city in cities:
    city_dict = {}
    city_dict['name'] = city
    if '/' in city:
        city = city.split('/')[1]
    url = f'http://api.openweathermap.org/geo/1.0/direct?q={city}&limit={1}&appid={api_key}'
    response = requests.get(url).json()
    city_dict['lat'] = response[0]['lat']
    city_dict['lon'] = response[0]['lon']
    city_coords.append(city_dict)
city_coords
```

```python
city_lats = {}
city_lons = {}
for city in city_coords:
    city_lats[f"{city['name']}"] = city['lat']
    city_lons[f"{city['name']}"] = city['lon']
```

```python
city_lons
```

```python
origin_lats = []
origin_lons = []
dest_lats = []
dest_lons = []
for row in df.iterrows():
    origin_lats.append(city_lats[f"{row[1]['ORIGIN_CITY_NAME']}"])
    origin_lons.append(city_lons[f"{row[1]['ORIGIN_CITY_NAME']}"])
    dest_lats.append(city_lats[f"{row[1]['DEST_CITY_NAME']}"])
    dest_lons.append(city_lons[f"{row[1]['DEST_CITY_NAME']}"])
df['DEST_LAT'] = dest_lats
df['DEST_LON'] = dest_lons
df['ORIGIN_LAT'] = origin_lats
df['ORIGIN_LON'] = origin_lons
```

```python
df
```

```python
df.to_csv('../Resources/MCI_21-23.csv')
```
